```python
## import library yang akan digunakan
from sqlalchemy import create_engine
from urllib.parse import quote_plus
import pandas as pd
```

```python
data = pd.read_csv('aircraft_dataset.csv')
data.head()
```

```python
# drop column Unnamed: 0
data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
## rename column 
new_columns = { 'Aircraft Model':'Aircraft_Model', 'Empty Weight (kg)':'Empty_Weight',
                'Maximum Takeoff Weight (kg)': 'Maximum_Takeoff_Weight', 'Maximum Landing Weight (kg)': 'Maximum_Landing_Weight', 
                'Center of Gravity Limits':'Center_Gravity_Limits', 'Wing Loading (kg/m²)':'Wing_Loading', 
                'Tail Loading (kg/m²)':'Tail_Loading','Fuel Capacity (liters)':'Fuel_Capacity', 
                'Number of Passengers':'Maximum_Passengers', 'Cargo Capacity (kg)':'Cargo_Capacity',
                'Maximum Altitude (ft)':'Maximum_Altitude', 'Maximum Speed (knots)':'Maximum_Speed', 
                'Engine Type':'Engine_Type', 'Fuselage Length (m)':'Fuselage_Length'}
```

```python
data.rename(columns=new_columns, inplace=True)
```

```python
## import CSV to SQL using Sqlalchemy
encoded_pass = quote_plus('PASSWORD') ## SQL Password
mysql_engine = create_engine(f"mysql://root:{encoded_pass}@localhost:3306/aircraft")
```

```python
## import CSV to Mysql
data.to_sql('aircraft_table', mysql_engine, if_exists='replace', index=False)
```
